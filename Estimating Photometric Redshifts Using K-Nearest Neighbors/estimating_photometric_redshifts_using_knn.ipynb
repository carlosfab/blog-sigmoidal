## Como funciona o k-NN

O k-Nearest Neighbor (k-NN) é reconhecido como um dos algoritmos de classificação mais simples e fundamentais em machine learning. A ideia central por trás do k-NN é que ele não "aprende" de fato no sentido tradicional. Em vez disso, opera com base na premissa de que os pontos de dados similares tendem a estar próximos uns dos outros no espaço de características. Isso significa que o k-NN depende diretamente da distância entre os vetores de características para realizar suas predições.

O cenário estatístico é fundamental para entender como o algoritmo opera sobre os dados. Suponha que temos pares $(X_1, Y_1), (X_2, Y_2), \dots, (X_n, Y_n)$ tomando valores em $\mathbb{R}^d \times \{1, 2\}$. Aqui, $X$ representa as características ou atributos dos pontos de dados em um espaço d-dimensional, e $Y$ é o rótulo da classe de $X$, assumindo valores em $\{1, 2\}$, indicando a qual das duas classes o ponto pertence.

Cada $X$ condicional a $Y=r$ segue uma distribuição de probabilidade $P_r$ para $r=1, 2$. Isso significa que, dado um rótulo de classe específico, a distribuição dos pontos de dados em $X$ segue um padrão específico, descrito pela distribuição \(P_r\).

Dada uma norma $\|\cdot\|$ em $\mathbb{R}^d$ e um ponto $x \in \mathbb{R}^d$, ordenamos os dados de treinamento tal que $(X_{(1)}, Y_{(1)}), \dots, (X_{(n)}, Y_{(n)})$ de forma que $\|X_{(1)} - x\| \leq \dots \leq \|X_{(n)} - x\|$. Em outras palavras, reorganizamos os dados de treinamento com base na proximidade de cada ponto $X_i$ ao ponto de consulta $x$, do mais próximo ao mais distante.

Imagine o k-NN como um motor de busca de imagens, onde a busca é realizada não por palavras-chave, mas pela semelhança visual entre as imagens. De maneira análoga, o k-NN classifica pontos de dados desconhecidos encontrando a classe mais comum entre os k exemplos mais próximos. Cada ponto de dados nos k exemplos mais próximos dá um "voto", e a categoria com mais votos é a classificação atribuída ao ponto em questão.

Considere, por exemplo, um conjunto de dados representando a 'Fluffiness' e a 'Leveza' do pelo de diferentes animais. O k-NN ajudaria a classificar um animal desconhecido com base na proximidade de seus atributos aos de animais conhecidos.

### Métricas de Distância

A escolha da métrica de distância é crucial e pode variar dependendo da natureza e dimensionalidade dos dados. Essas métricas são fundamentais para determinar quão "próximos" ou "semelhantes" são dois pontos de dados no espaço de características. Aqui estão detalhes mais matemáticos de algumas métricas comuns:

**Distância Euclidiana:** A mais comum e intuitiva entre as métricas, usada para medir a distância linear entre dois pontos. Se tivermos dois pontos, $P = (p_1, p_2, ..., p_n)$ e $Q = (q_1, q_2, ..., q_n)$ em um espaço $n$-dimensional, a distância Euclidiana entre eles é dada por:

  $$
  d(P, Q) = \sqrt{(p_1 - q_1)^2 + (p_2 - q_2)^2 + \cdots + (p_n - q_n)^2}
  $$

É particularmente útil quando as imagens ou pontos de dados são representados em um espaço euclidiano, fornecendo uma medida direta da "linha reta" entre eles.

**Distância Manhattan (city block):** Também conhecida como norma L1, esta métrica mede a distância entre dois pontos se movendo apenas em linhas retas ao longo dos eixos (como um táxi se movendo em uma grade de ruas da cidade). Para os mesmos pontos \(P\) e \(Q\) acima, a distância Manhattan é calculada como:

  $$
  d(P, Q) = |p_1 - q_1| + |p_2 - q_2| + \cdots + |p_n - q_n|
  $$

Esta métrica é ideal para quando o caminho entre pontos é uma grade, como em um cenário urbano ou em alguns tipos de gráficos.

**Outras métricas:** Dependendo do tipo de dados e do problema, outras métricas de distância podem ser mais apropriadas. Por exemplo:
- **Distância Qui-quadrado:** Frequentemente usada para comparar histogramas e distribuições de probabilidade, calculada como $\sum \frac{(O_i - E_i)^2}{E_i}$, onde $O_i$ e $E_i$ são os valores observados e esperados, respectivamente.
- **Distância de Minkowski:** Uma generalização das distâncias Euclidiana e Manhattan, definida como $ (\sum{|p_i - q_i|^r})^{1/r} $, onde $r$ é um parâmetro que determina a natureza da distância.

Cada métrica tem suas peculiaridades e pode ser mais adequada para certos tipos de dados ou problemas. A escolha da métrica de distância afeta significativamente o desempenho do classificador k-NN, e entender suas propriedades matemáticas pode ajudar a fazer uma escolha mais informada.


### Aplicações e Implicações

O k-NN é ideal para sistemas simples de classificação e reconhecimento de imagens, onde a rapidez e a simplicidade são mais valiosas do que a precisão perfeita. As categorias são definidas pela proximidade e semelhança dos dados, o que significa que a escolha da métrica de distância e o valor de k podem afetar significativamente o desempenho.

Os dados em um problema k-NN são frequentemente visualizados como agrupados em um espaço n-dimensional baseado em suas características. Neste espaço, a proximidade entre pontos indica a semelhança. Isso não só ajuda a entender a distribuição dos dados, mas também fornece insights sobre como diferentes características contribuem para a classificação.

Em resumo, o k-NN, apesar de sua simplicidade, é uma ferramenta poderosa em certos contextos, oferecendo uma abordagem direta e intuitiva para classificação e análise de dados. Ao entender seus fundamentos e aplicações, os praticantes podem fazer uso eficiente dessa técnica e reconhecer quando ela é a escolha mais apropriada para o problema em mãos.


````python
# Importando as bibliotecas necessárias
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

# Gerando um conjunto de dados aleatórios
np.random.seed(0)
X = np.random.rand(100, 2)  # 100 pontos em 2 dimensões
y = np.where(X[:, 0] + X[:, 1] > 1, 1, 0)  # Classificação baseada na soma das características

# Visualizando os dados
plt.figure(figsize=(8, 6))
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='red', label='Classe 0')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='blue', label='Classe 1')
plt.title('Conjunto de Dados Gerado')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
````

Após a geração e visualização dos dados, o próximo passo é treinar o modelo k-NN. Utilizaremos a biblioteca Scikit-learn para criar e treinar o modelo.


````python
# Definindo o número de vizinhos
k = 3

# Criando o modelo k-NN
knn = KNeighborsClassifier(n_neighbors=k)

# Treinando o modelo com os dados gerados
knn.fit(X, y)
````

Com o modelo treinado, podemos prever a classificação de novos pontos e visualizar as fronteiras de decisão criadas pelo k-NN.

````python
# Gerando pontos de teste para visualização da fronteira de decisão
x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Visualizando a fronteira de decisão
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.4)
plt.scatter(X[:, 0], X[:, 1], c=y, s=20, edgecolor='k')
plt.title('Fronteira de decisão do k-NN')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
````

Este exemplo básico demonstra a implementação e a aplicação do k-NN em um conjunto de dados sintéticos. A fronteira de decisão visualizada ajuda a entender como o algoritmo classifica novos dados baseados em seus vizinhos mais próximos.

Esta consulta SQL seleciona as primeiras 50.000 entradas de galáxias do SDSS combinando dados fotométricos de `PhotoObj` (p) e dados espectroscópicos de `SpecObj` (s). Ela filtra galáxias com magnitudes u e g dentro de intervalos específicos e exclui certas classes de objetos como estrelas e céu. Os resultados, incluindo magnitudes u, g, r, i, z, classe espectral, redshift e erro de redshift, são salvos em `MyTable` no banco de dados `mydb`.

```sql
SELECT TOP 50000
  p.u, p.g, p.r, p.i, p.z, s.class, s.z, s.zerr into mydb.MyTable from PhotoObj AS p
  JOIN SpecObj AS s ON s.bestobjid = p.objid
WHERE 
  p.u BETWEEN 0 AND 19.6
  AND p.g BETWEEN 0 AND 20
  AND s.class <> 'UNKNOWN'
  AND s.class <> 'STAR'
  AND s.class <> 'SKY'
  AND s.class <> 'STAR_LATE'
```

````python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

from sklearn.neighbors import KNeighborsRegressor

# Assuming astroML is already installed and available
from astroML.plotting import scatter_contour

n_neighbors = 1

# Load the data
data = pd.read_csv("data/sdss_galaxy_colors_top50k.csv")

# Shuffle data
np.random.seed(0)
shuffled_indices = np.random.permutation(len(data))
data = data.iloc[shuffled_indices]

N = len(data)

# Put colors in a matrix
X = np.zeros((N, 4))
X[:, 0] = data['u'] - data['g']
X[:, 1] = data['g'] - data['r']
X[:, 2] = data['r'] - data['i']
X[:, 3] = data['i'] - data['z']

# Assuming 'Column1' is the target variable you want to predict
y = data['Column1'].values  # Ensure this is a NumPy array for consistency

# Divide into training and testing data
Ntrain = N // 2
Xtrain = X[:Ntrain]
ytrain = y[:Ntrain]

Xtest = X[Ntrain:]
ytest = y[Ntrain:]

# Initialize and train the k-NN regressor
knn = KNeighborsRegressor(n_neighbors, weights='uniform')
knn.fit(Xtrain, ytrain)
ypred = knn.predict(Xtest)

# Assess the model (you can choose an appropriate metric depending on your problem)
# Example: Root Mean Square Error (RMSE)
rms = np.sqrt(np.mean((ytest - ypred) ** 2))
print(f"RMS error = {rms:.2g}")
````

````python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.neighbors import KNeighborsRegressor

# Assuming astroML is already installed and available
# from astroML.plotting import scatter_contour

# Definindo as configurações do gráfico
def plot_results(ytest, ypred, rms):
    # axis_lim = np.array([ytest.min() - 0.1, ytest.max() + 0.1])
    axis_lim = np.array([0, 2.5])

    plt.figure(figsize=(10, 6))
    ax = plt.axes()
    plt.scatter(ytest, ypred, c='k', lw=0, s=4)
    plt.plot(axis_lim, axis_lim, '--k')
    plt.plot(axis_lim, axis_lim + rms, ':k')
    plt.plot(axis_lim, axis_lim - rms, ':k')
    plt.xlim(axis_lim)
    plt.ylim(axis_lim)

    plt.text(0.98, 0.02, f"RMS error = {rms:.2g}", ha='right', va='bottom', transform=ax.transAxes,
             bbox=dict(ec='w', fc='w'), fontsize=12)

    plt.title('Prediction Comparison')
    plt.xlabel(r'$\mathrm{Actual}$', fontsize=14)
    plt.ylabel(r'$\mathrm{Predicted}$', fontsize=14)
    plt.show()

# Dividindo os dados
Ntrain = N // 2
Xtrain = X[:Ntrain]
ytrain = y[:Ntrain]
Xtest = X[Ntrain:]
ytest = y[Ntrain:]

# Treinando o modelo
knn = KNeighborsRegressor(n_neighbors=1, weights='uniform')
knn.fit(Xtrain, ytrain)
ypred = knn.predict(Xtest)

# Calculando o erro RMS
rms = np.sqrt(np.mean((ytest - ypred) ** 2))

# Criando o gráfico
plot_results(ytest, ypred, rms)
````

````python
def plot_improved_results(ytest, ypred, rms):
    # Definindo limites do eixo
    axis_lim = np.array([0, 2.5])

    plt.figure(figsize=(10, 6))
    ax = plt.axes()

    # Utilizando uma escala de cores baseada no erro absoluto de cada ponto
    errors = np.abs(ytest - ypred)
    scatter = plt.scatter(ytest, ypred, c=errors, cmap='coolwarm', lw=0, s=4, alpha=0.6)
    plt.colorbar(scatter, label='Erro Absoluto')

    # Linhas para representar o erro RMS
    plt.plot(axis_lim, axis_lim, '--b', label='Ideal')
    plt.plot(axis_lim, axis_lim + rms, ':r', label=f'+ RMS ({rms:.2f})')
    plt.plot(axis_lim, axis_lim - rms, ':r', label=f'- RMS ({rms:.2f})')

    # Configurando limites, grade e legendas
    plt.xlim(axis_lim)
    plt.ylim(axis_lim)
    plt.grid(True, which='both', ls='--', lw=0.5)
    plt.legend()

    # Anotações e títulos
    plt.text(0.98, 0.02, f"RMS error = {rms:.2g}", ha='right', va='bottom', transform=ax.transAxes,
             bbox=dict(ec='w', fc='w'), fontsize=12)

    plt.title('Photo-z: Nearest Neighbor Regression Comparison')
    plt.xlabel(r'$\mathrm{z_{spec}}$', fontsize=14)
    plt.ylabel(r'$\mathrm{z_{phot}}$', fontsize=14)
    plt.show()

# Criando o gráfico melhorado
plot_improved_results(ytest, ypred, rms)
````
